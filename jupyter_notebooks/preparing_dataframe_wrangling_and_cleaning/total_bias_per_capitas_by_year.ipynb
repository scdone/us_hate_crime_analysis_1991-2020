```python
pop_list = df.population_estimate.tolist() # sets population estimates to list 

# function to find per capita incidents

def per_capita_func(y):
    x = []
    
    for pop, total in zip(pop_list, y):
        x.append((total/pop)*100000)
        
    return x
```

```python
anti_american_indian_list = df.anti_american_indian_or_alaska_native.tolist()
anti_american_indian_or_alaska_native_per_capita = per_capita_func(anti_american_indian_list) 
df.insert(3, 'anti_american_indian_or_alaska_native_per_capita', anti_american_indian_or_alaska_native_per_capita)
df.drop(columns=['anti_american_indian_or_alaska_native'], inplace=True)
```

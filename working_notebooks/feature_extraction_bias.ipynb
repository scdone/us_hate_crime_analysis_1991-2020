```python
conditions = [
    (incidents['bias_grouping'] == "Anti-LGBTQ"),
    (incidents['bias_grouping'] != "Anti-LGBTQ")
]

values = [True, False]

incidents['anti_lgbtq_grouped'] = np.select(conditions,values)
```

```python
incidents.drop(columns=["bias_grouping"], inplace=True)
```

```python
column_list = incidents.columns.tolist()
```

```python
# code solution to change all to snake_case found here - https://stackoverflow.com/a/1176023

import re

snake_list = []

for column in column_list:
    column = re.sub(r'(?<!^)(?=[A-Z])', '_', column).lower()
    column = column.replace("(","")
    column = column.replace(")","")
    column = column.replace("'","")
    column = column.replace(",","")
    column = column.replace("-","")
    column = column.replace(" ", "_")
    column = column.replace("/", "")
    column = column.replace("__", "_")
        
    snake_list.append(column)
```

```python
incidents.columns = snake_list
```

```python
pop_list = per_capita.population_estimate.tolist()
```

```python
total_incidents_list = per_capita.total_incidents.tolist()
```

```python
per_capita_list = []

def per_capita_func():
    for pop, total in zip(pop_list, total_incidents_list):

        per_capita_list.append((total/pop)*100000)
            
```
